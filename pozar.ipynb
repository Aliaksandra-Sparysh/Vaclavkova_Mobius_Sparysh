### Požáry v Řecku, 2023

V červenci 2023 zasáhla Řecko série ničivých lesních požárů, které si vyžádaly nejméně 28 obětí a 75 zraněných. Celkem bylo zaznamenáno přes 80 jednotlivých požárů, přičemž 79 osob bylo zatčeno za žhářství.

Teploty v zemi dosahovaly 41 °C, přičemž na ostrově Rhodos meteorologové předpovídali maxima kolem 45 °C. Odborníci varovali, že se jedná o nejdelší vlnu veder v dějinách Řecka, která měla trvat až 16–17 dní. Po sérii veder a požárů napříč Evropou vypukly první ohně v Řecku 17. července. Vláda zareagovala vytvořením Krizového řídicího štábu, který koordinoval zásahy. S postupujícím rozsahem škod a nárůstem spálené půdy začali obyvatelé kritizovat nedostatečné kapacity hasičského sboru a obviňovali vládu z špatného řízení krizové situace.


### Požár na ostrově Rhodos

Ostrov Rhodos byl v červenci zasažen rozsáhlými lesními požáry, které si vyžádaly evakuaci přibližně 19 000 lidí.
Podle místního hasičského sboru šlo o největší evakuační operaci v historii ostrova.

20. července byl na Rhodosu vyhlášen stav nouze ve třech obcích, včetně Lindosu, který je známým turistickým centrem. 
21. července, nařídila civilní ochrana evakuaci vesnice Laerma, vzdálené asi 10 km od pobřežních letovisek. 
23. července bylo rozhodnuto o dalším přesunu 1 200 osob z obcí Pefki, Lindos a Kalathos.
